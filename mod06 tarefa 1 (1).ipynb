```python
# Importando as bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

# Importando os dados
df_sus = pd.read_csv("./datasus_nascidos_vivos.csv")
df_sus
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
# Criando o eixo e a figura
fig, ax = plt.subplots()

# Plotando os dados da varivel
ax.hist(df_sus["APGAR1"])
plt.show()
```

```python
# Criando o eixo e a figura
fig, ax = plt.subplots()

# Plotando os dados da viriavel
ax.hist(df_sus["APGAR5"])
plt.show
```

```python
# Criando o eixo, figura e plotando o gráfico com o metodo countplot()
sns.countplot(data=df_sus, x="SEXO")
plt.show()
```

```python
# Criando o eixo e a figura
fig, ax = plt.subplots()

# Plotando os dados da varivel
ax = sns.countplot(data=df_sus, x="GESTACAO")

# Modificando a rotacao dos ticks
ax.tick_params(axis="x", rotation=90)
```

```python
# Criando o eixo e a figura
fig, ax = plt.subplots()

# Plotando os dados da viriavel
ax = sns.countplot(data=df_sus, x="ESTCIVMAE")

# Alterando e rotacionando os valores dos ticks
labels = ax.set_xticklabels(["Solteira", "Casada", "Viuva", "Separada", "NAN", "Ignorado"], rotation=45)
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
# Criando o eixo e a figura
fig, ax = plt.subplots()

# Plotando os dados da variavel
ax = sns.barplot(data=df_sus, x="APGAR5", y="CONSULTAS")
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
# Criando o eixo e a figura
fig, ax = plt.subplots(2,2)

# Modificando tamanho e espassamento
plt.rc("figure", figsize=(10, 8))
plt.subplots_adjust(wspace=0.2, hspace=0.6)

# Plotando os 4 graficos
ax1 = sns.countplot(data=df_sus, x="SEXO", ax=ax[0,0])
ax2 = sns.countplot(data=df_sus, x="GESTACAO", ax=ax[0,1])
ax3 = sns.countplot(data=df_sus, x="ESTCIVMAE", ax=ax[1,0])
ax4 = sns.countplot(data=df_sus, x="CONSULTAS", ax=ax[1,1])

# Modificacoes estéticas
ax2.tick_params(rotation=30)
ax3.set_xticklabels(["Solteira", "Casada", "Viuva", "Separada", "NONE", "Ignorado"])
ax4.set_xticklabels(["Nenhuma", "de 1 a 3", "de 4 a 6", "7 e mais", "Nenhuma"])
plt.show()
```
